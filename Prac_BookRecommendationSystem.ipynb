##3.Popularity Based Recommender System

Here to get the most popular book we use the formula of average ratings.We will only consider books having number of ratings greater tha 250.


```python
ratings_with_name=ratings.merge(books,on="ISBN")
```

```python
ratings_with_name.head()
```

```python
No_of_ratings=ratings_with_name.groupby("Book-Title").count()["Book-Rating"].reset_index()
No_of_ratings.head()
```

```python
No_of_ratings=No_of_ratings.rename(columns={"Book-Rating":"No_Of_Ratings"})
```

```python
No_of_ratings.head()
```

```python
Average_ratings=ratings_with_name.groupby("Book-Title").mean()["Book-Rating"].reset_index()
Average_ratings.head()
```

```python
Average_ratings=Average_ratings.rename(columns={"Book-Rating":"Average_Ratings"})
Average_ratings
```

```python
Popularity_DF=No_of_ratings.merge(Average_ratings,on="Book-Title")
Popularity_DF.head()
```

```python
Popularity_DF=Popularity_DF[Popularity_DF["No_Of_Ratings"]>=250].sort_values("Average_Ratings",ascending=False).head(50)
```

```python
Popularity_DF
```

```python
Popularity_DF=Popularity_DF.merge(books,on="Book-Title").drop_duplicates("Book-Title")[["Book-Title","Book-Author","Image-URL-M","No_Of_Ratings","Average_Ratings"]]
```

```python
Popularity_DF
```

##3.Collaborative Filtering Based Recommender System
Here we will be taking books which are rated by more than 200 users and those users who have rated more than 50 books.

```python
ratings_with_name.groupby("User-ID").count()["Book-Rating"]
```

out of total users only 92106 users have rated any book.

```python
X = ratings_with_name.groupby("User-ID").count()["Book-Rating"]>200
X[X]
```

```python
Frequent_Readers = X[X].index
Frequent_Readers
```

```python
Filtered_Ratings = ratings_with_name[ratings_with_name["User-ID"].isin(Frequent_Readers)]
Filtered_Ratings
```

```python
Popularity_DF["Image-URL-M"][0]
```

out of total users only 474007 users were Frequent users who have rated the books

```python
y = Filtered_Ratings.groupby("Book-Title").count()["Book-Rating"]>=50
Famous_Books = y[y].index
```

```python
Final_Ratings = Filtered_Ratings[Filtered_Ratings["Book-Title"].isin(Famous_Books)]
```

we have 58586 famous books

```python
# Pivot TAble

Pivot_table = Final_Ratings.pivot_table(index="Book-Title",columns="User-ID",values="Book-Rating")
```

```python
Pivot_table.fillna(0,inplace=True)
Pivot_table
```

we can imagine each book as a point in 810 dimentional system and we will find the eucledian distance to find the similar books.

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
Similarity_Score = cosine_similarity(Pivot_table)
```

```python
Similarity_Score
```

first array is discribing the distance of the first book(1984) with other books

```python
list(enumerate(Similarity_Score[0]))  # it gives the similiarity score of the book 1984 with all other books
```

```python
# to get top 5  sorted according to the similarity in ascending way
sorted(list(enumerate(Similarity_Score[0])),key=lambda x:x[1],reverse=True)[1:6]
```

```python
## creating the function
def recommend(book_name):
  index = np.where(Pivot_table.index==book_name)[0][0]
  similar_items = sorted(list(enumerate(Similarity_Score[index])),key=lambda x:x[1],reverse=True)[1:6]

  for i in similar_items:
    print(Pivot_table.index[i[0]])
```

```python
# Test it for roamntic Novels
recommend("Message in a Bottle")
```

```python
recommend("The Notebook")
```
